## 1.1 Warum Versionskontrolle?

In der Energiewirtschaft arbeiten wir oft mit:

**Mit Git:**
- Jede Г„nderung ist nachvollziehbar
- Man kann jederzeit zu einem frГјheren Stand zurГјckkehren
- Mehrere Personen kГ¶nnen gleichzeitig arbeiten
- Code-Reviews und QualitГӨtskontrolle werden mГ¶glich

### Wichtige Begriffe

| Begriff | Beschreibung |
|---------|-------------|
| **Repository (Repo)** | Ein Projektordner, der von Git verwaltet wird |
| **Commit** | Ein Snapshot aller getrackten Dateien zu einem Zeitpunkt |
| **Remote** | Ein entferntes Repository (z.B. auf GitHub) |
| **Clone** | Kopie eines Remote-Repos auf deinen Rechner |
| **Push** | Lokale Commits zu GitHub hochladen |
| **Pull** | Г„nderungen von GitHub herunterladen und integrieren |

## 1.4 Der Standard-Workflow

### Schritt 1: Repository von GitHub klonen

Erstelle zuerst ein **neues leeres Repository auf GitHub** (im Browser, mit README вң…).  
Dann einmalig klonen вҖ” damit ist die Verbindung zu GitHub automatisch gesetzt:

```bash
git clone https://github.com/dein-user/dein-repo.git
cd dein-repo
```

### Schritt 2: TГӨgliche Arbeit

```bash
# 1. Status prГјfen вҖ“ Was hat sich geГӨndert?
git status

# 2. Dateien zur Staging Area hinzufГјgen
git add dateiname.py        # Einzelne Datei
git add .                   # Alles im aktuellen Verzeichnis

# 3. Commit erstellen (Snapshot speichern)
git commit -m "Batterie-Optimierung: Wirkungsgrad hinzugefГјgt"

# 4. Г„nderungen zu GitHub hochladen
git push

# 5. Г„nderungen von GitHub herunterladen (notwendig, wenn andere mitarbeiten und andere Personen Г„nderungen gepusht haben)
git pull
```

> рҹ’Ў **Tipp fГјr Commit-Messages:** Beschreibe WAS und WARUM, nicht WIE.
>
> вң… `"Fix: Negative Preise fГјhrten zu falschem Batterie-SoC"`
>
> вқҢ `"Г„nderung"` oder `"Update"`

## Add-on B: Branches (Verzweigungen)

Branches erlauben es, neue Features zu entwickeln, ohne den funktionierenden Code auf `main` zu gefГӨhrden.

```bash
# Neuen Branch erstellen und wechseln
git switch -c feature/pv-anlage

# Zwischen Branches wechseln
git switch main

# Branch zusammenfГјhren
git switch main
git merge feature/pv-anlage

# Branch lГ¶schen (nach erfolgreichem Merge)
git branch -d feature/pv-anlage
```

**Typische Branch-Strategie:**
```
main в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв—Ҹв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв—Ҹв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв—Ҹв”Җв”Җв”Җв”Җв”Җв–ә
                \          /
 feature/        в—Ҹв”Җв”Җв”Җв”Җв—Ҹв”Җв”Җв”Җв—Ҹ
 batterie-opt
```

### Гңbung

```bash
git switch -c feature/co2-berechnung
# FГјgt eine COвӮӮ-Methode zur Kraftwerk-Klasse hinzu
git add kraftwerk.py
git commit -m "Feature: COвӮӮ-Berechnung hinzugefГјgt"
git switch main
git merge feature/co2-berechnung
git push
```

## Add-on C: Team-Workflow mit Pull Requests

So arbeitet man professionell im Team:

```
1. git pull                                  вҶҗ Aktuelle Version holen
2. git switch -c feature/mein-feature        вҶҗ Neuen Branch erstellen
3. ... Dateien bearbeiten ...
4. git add .                                 вҶҗ Г„nderungen vormerken
5. git commit -m "Beschreibung"              вҶҗ Commit erstellen
6. git push -u origin feature/mein-feature   вҶҗ Branch zu GitHub hochladen
7. Pull Request auf GitHub erstellen         вҶҗ Im Browser
8. Code Review durch Kolleg:innen
9. Merge in main
```

> рҹ’Ў Ein **Pull Request (PR)** ist eine Anfrage, den eigenen Branch in `main` zu mergen.
> Dort kГ¶nnen Kolleg:innen Kommentare hinterlassen, bevor der Code integriert wird.

## Add-on D: Fehler rГјckgГӨngig machen

```bash
# Г„nderungen an einer Datei verwerfen (vor git add)
git restore dateiname.py

# Datei aus Staging Area entfernen (nach git add, vor commit)
git restore --staged dateiname.py

# Letzten Commit rГјckgГӨngig machen (Г„nderungen bleiben erhalten)
git reset --soft HEAD~1

# Letzten Commit komplett rГјckgГӨngig machen
# вҡ пёҸ Vorsicht: Г„nderungen sind danach weg!
git reset --hard HEAD~1
```

## Add-on E: Historie & Г„nderungen im Detail

```bash
# Г„nderungen anzeigen (vor git add)
git diff

# Commit-Historie kompakt
git log --oneline

# Mit Branch-Visualisierung
git log --oneline --graph --all

# Г„nderungen eines bestimmten Commits anzeigen
git show abc1234

# Wer hat welche Zeile zuletzt geГӨndert?
git blame dateiname.py

# Г„nderungen zwischenspeichern (ohne Commit)
git stash
git stash pop       # Wieder herstellen
```
