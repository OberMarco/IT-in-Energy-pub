## 1.1 Warum Versionskontrolle?

In der Energiewirtschaft arbeiten wir oft mit:
- Optimierungsmodellen (z.B. Einsatzplanung Kraftwerke)
- Datenanalyse-Skripten (Lastprofile, Preisprognosen)
- Konfigurationsdateien (Anlagenparameter)

**Probleme ohne Versionskontrolle:**
```
optimierung_v1.py
optimierung_v2.py
optimierung_v2_final.py
optimierung_v2_final_wirklich.py
optimierung_v2_final_korrektur_chef.py    ← 😩
```

**Mit Git:**
- Jede Änderung ist nachvollziehbar
- Man kann jederzeit zu einem früheren Stand zurückkehren
- Mehrere Personen können gleichzeitig arbeiten
- Code-Reviews und Qualitätskontrolle werden möglich

## 1.2 Git-Konzepte

### Die drei Bereiche von Git

```
┌─────────────────┐     git add      ┌─────────────────┐    git commit    ┌─────────────────┐
│  Working        │ ──────────────►  │  Staging Area   │ ──────────────►  │  Repository     │
│  Directory      │                  │  (Index)        │                  │  (.git/)        │
│                 │                  │                 │                  │                 │
│  Deine Dateien  │                  │  Vorgemerkte    │                  │  Gespeicherte   │
│  die du         │                  │  Änderungen     │                  │  Historie       │
│  bearbeitest    │                  │                 │                  │                 │
└─────────────────┘                  └─────────────────┘                  └─────────────────┘
```

### Wichtige Begriffe

| Begriff | Beschreibung |
|---------|-------------|
| **Repository (Repo)** | Ein Projektordner, der von Git verwaltet wird |
| **Commit** | Ein Snapshot aller getrackte Dateien zu einem Zeitpunkt |
| **Branch** | Ein unabhängiger Entwicklungszweig |
| **Remote** | Ein entferntes Repository (z.B. auf GitHub/GitLab) |
| **Clone** | Kopie eines Remote-Repos auf deinen Rechner |
| **Push** | Lokale Commits zum Remote hochladen |
| **Pull** | Änderungen vom Remote herunterladen und integrieren |

## 1.3 Die wichtigsten Git-Befehle

### Einmalige Konfiguration

```bash
# Name und E-Mail setzen (wird in jedem Commit gespeichert)
git config --global user.name "Max Mustermann"
git config --global user.email "max.mustermann@stadtwerk.de"

# Standard-Editor setzen (optional)
git config --global core.editor "code --wait"

# Konfiguration anzeigen
git config --list
```

### Repository erstellen oder klonen

```bash
# Neues Repository in aktuellem Ordner erstellen
git init

# Bestehendes Repository klonen
git clone https://github.com/user/projekt.git

# Klonen in bestimmten Ordner
git clone https://github.com/user/projekt.git mein-ordner
```

### Tägliche Arbeit: Der Standard-Workflow

```bash
# 1. Status prüfen – Was hat sich geändert?
git status

# 2. Änderungen anzeigen (im Detail)
git diff

# 3. Dateien zur Staging Area hinzufügen
git add dateiname.py              # Einzelne Datei
git add ordner/                    # Ganzer Ordner
git add .                          # Alles im aktuellen Verzeichnis
git add *.py                       # Alle Python-Dateien

# 4. Commit erstellen (Snapshot speichern)
git commit -m "Batterie-Optimierung: Wirkungsgrad hinzugefügt"

# 5. Änderungen zum Remote-Repository hochladen
git push

# 6. Änderungen vom Remote-Repository herunterladen
git pull
```

> 💡 **Tipp für Commit-Messages:** Beschreibe WAS und WARUM, nicht WIE.
>
> ✅ `"Fix: Negative Preise führten zu falschem Batterie-SoC"`
>
> ❌ `"Änderung"` oder `"Update"`

### Historie anschauen

```bash
# Commit-Historie anzeigen
git log

# Kompakte Ansicht (eine Zeile pro Commit)
git log --oneline

# Mit Graphen-Ansicht (Branches visualisieren)
git log --oneline --graph --all

# Änderungen eines bestimmten Commits anzeigen
git show abc1234

# Wer hat welche Zeile geschrieben?
git blame dateiname.py
```

### Branches (Verzweigungen)

```bash
# Alle Branches anzeigen
git branch

# Neuen Branch erstellen und wechseln
git checkout -b feature/pv-anlage
# Oder in neueren Git-Versionen:
git switch -c feature/pv-anlage

# Zwischen Branches wechseln
git checkout main
git switch main

# Branch zusammenführen (merge)
git checkout main          # Erst zum Ziel-Branch wechseln
git merge feature/pv-anlage

# Branch löschen (nach erfolgreichem Merge)
git branch -d feature/pv-anlage
```

**Typische Branch-Strategie in Energie-Projekten:**
```
main ──────────●────────────●────────────●─────►
                \          /              \
 feature/       ●────●───●                \
 batterie-opt                              \
                                   feature/ ●───●
                                   regression
```

### Fehler rückgängig machen

```bash
# Änderungen an einer Datei verwerfen (vor git add)
git checkout -- dateiname.py
# Neuere Syntax:
git restore dateiname.py

# Datei aus Staging Area entfernen (nach git add, vor commit)
git reset HEAD dateiname.py
# Neuere Syntax:
git restore --staged dateiname.py

# Letzten Commit rückgängig machen (Änderungen bleiben erhalten)
git reset --soft HEAD~1

# Letzten Commit komplett rückgängig machen (⚠️ Änderungen weg!)
git reset --hard HEAD~1
```

### .gitignore – Dateien ausschließen

Erstelle eine Datei `.gitignore` im Hauptverzeichnis:

```gitignore
# Python
__pycache__/
*.pyc
.ipynb_checkpoints/
venv/
.env

# Daten (große Dateien nicht in Git!)
*.csv
*.xlsx
data/raw/

# IDE
.vscode/
.idea/

# Outputs
outputs/
*.html
```

> 💡 **Warum?** Große Datendateien (Lastprofile, Wetterdaten) gehören **nicht** in Git.
> Nutzt dafür separate Datenspeicher oder Git LFS.

## 1.4 Remote-Repositories (GitHub / GitLab)

```bash
# Remote-Repository hinzufügen
git remote add origin https://github.com/user/projekt.git

# Remote-Repos anzeigen
git remote -v

# Ersten Push (mit Upstream-Tracking)
git push -u origin main

# Danach reicht:
git push
git pull
```

### Typischer Workflow im Team

```
1. git pull                              ← Aktuelle Version holen
2. git checkout -b feature/mein-feature  ← Neuen Branch erstellen
3. ... Dateien bearbeiten ...
4. git add .                             ← Änderungen vormerken
5. git commit -m "Beschreibung"          ← Commit erstellen
6. git push -u origin feature/mein-feat  ← Branch hochladen
7. Pull Request / Merge Request erstellen ← Im Browser (GitHub/GitLab)
8. Code Review durch Kolleg:innen
9. Merge in main
```

### Aufgabe: Erstes eigenes Repository

Öffnet ein **Terminal** und führt folgende Schritte durch:

1. **Erstellt einen neuen Ordner und initialisiert ein Git-Repository:**
   ```bash
   mkdir energiewirtschaft-uebung
   cd energiewirtschaft-uebung
   git init
   ```

2. **Erstellt eine Python-Datei `kraftwerk.py`** mit der Klasse aus Übung 1
   (copy-paste ist erlaubt 😉)

3. **Macht euren ersten Commit:**
   ```bash
   git add kraftwerk.py
   git commit -m "Initial: Kraftwerk-Klasse hinzugefügt"
   ```

4. **Überprüft die Historie:**
   ```bash
   git log --oneline
   ```

5. **Erstellt einen neuen Branch, ändert etwas und merged:**
   ```bash
   git checkout -b feature/co2-berechnung
   # Fügt die CO₂-Methode zur Klasse hinzu
   git add kraftwerk.py
   git commit -m "Feature: CO₂-Berechnung hinzugefügt"
   git checkout main
   git merge feature/co2-berechnung
   ```

6. **Bonus: Erstellt ein Repository auf GitHub und pusht euren Code:**
   ```bash
   git remote add origin https://github.com/<user>/energiewirtschaft-uebung.git
   git push -u origin main
   ```
