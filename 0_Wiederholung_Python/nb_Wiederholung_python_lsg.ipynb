## 0.2 Listen und Dictionaries

```python
# Liste: Stündliche Erzeugung eines Windparks (MW) über 8 Stunden
erzeugung_mw = [120.5, 135.2, 142.0, 138.7, 110.3, 95.8, 105.1, 128.4]

print(f"Erzeugung Stunde 1: {erzeugung_mw[0]} MW")
print(f"Maximale Erzeugung: {max(erzeugung_mw)} MW")
print(f"Minimale Erzeugung: {min(erzeugung_mw)} MW")
print(f"Anzahl Stunden: {len(erzeugung_mw)}")

# Dictionary: Kraftwerksportfolio
portfolio = {
    "Windpark Nordsee": {"typ": "Wind", "leistung_mw": 150, "baujahr": 2020},
    "Solarpark Süd":    {"typ": "Solar", "leistung_mw": 80, "baujahr": 2021},
    "BHKW Mitte":       {"typ": "Gas", "leistung_mw": 20, "baujahr": 2015},
    "Batterie Ost":     {"typ": "Speicher", "leistung_mw": 10, "baujahr": 2023},
}

# Zugriff auf verschachtelte Daten
print(f"Leistung Solarpark: {portfolio['Solarpark Süd']['leistung_mw']} MW")

# Alle Kraftwerke auflisten
for name, daten in portfolio.items():
    print(f"  {name}: {daten['leistung_mw']} MW ({daten['typ']})")

# Gesamtleistung berechnen
gesamt = sum(kw["leistung_mw"] for kw in portfolio.values())
print(f"\nGesamte installierte Leistung: {gesamt} MW")
```

## 0.4 Schleifen

```python
# for-Schleife: Tagesproduktion berechnen
stundenerzeugung = [120.5, 135.2, 142.0, 138.7, 110.3, 95.8, 105.1, 128.4]
tagesproduktion = 0

for stunde, leistung in enumerate(stundenerzeugung):
    energie = leistung * 1  # 1 Stunde → MWh
    tagesproduktion += energie
    print(f"  Stunde {stunde + 1}: {leistung:6.1f} MW → {energie:6.1f} MWh")

print(f"\nTagesproduktion gesamt: {tagesproduktion:.1f} MWh")

# %%
# while-Schleife: Batterie laden bis voll
kapazitaet_mwh = 100.0
ladeleistung_mw = 25.0
soc = 20.0  # State of Charge in MWh
stunde = 0

print("Ladevorgang:")
while soc < kapazitaet_mwh:
    soc = min(soc + ladeleistung_mw, kapazitaet_mwh)
    stunde += 1
    print(f"  Stunde {stunde}: SoC = {soc:.0f} MWh ({soc/kapazitaet_mwh*100:.0f}%)")

print(f"Batterie voll nach {stunde} Stunden")

# %%
# List Comprehension: Preise filtern
preise = [42.5, -3.2, 78.1, 120.4, 55.0, -12.1, 95.3, 67.8]
negative_preise = [p for p in preise if p < 0]
hohe_preise = [p for p in preise if p > 80]

print(f"Negative Preise: {negative_preise}")
print(f"Hohe Preise (>80): {hohe_preise}")
```

## 0.5 Funktionen

```python
def berechne_erloes(erzeugung_mw: list, preise_eur_mwh: list) -> float:
    """
    Berechnet den Erlös aus Erzeugung und Marktpreisen.

    Args:
        erzeugung_mw: Liste der stündlichen Erzeugung in MW
        preise_eur_mwh: Liste der stündlichen Preise in EUR/MWh

    Returns:
        Gesamterlös in EUR
    """
    if len(erzeugung_mw) != len(preise_eur_mwh):
        raise ValueError("Listen müssen gleich lang sein!")

    erloes = 0
    for e, p in zip(erzeugung_mw, preise_eur_mwh):
        erloes += e * p * 1  # 1h Zeitschritt

    return erloes


# Test
erzeugung = [100, 120, 80, 90]
preise = [50, 65, 40, 85]
ergebnis = berechne_erloes(erzeugung, preise)
print(f"Erlös: {ergebnis:,.2f} EUR")


def volllaststunden(jahreserzeugung_mwh: float, installierte_leistung_mw: float) -> float:
    """Berechnet die Volllaststunden eines Kraftwerks."""
    return jahreserzeugung_mwh / installierte_leistung_mw


# Typische Werte
vlh_wind_onshore = volllaststunden(500_000, 250)
vlh_solar = volllaststunden(80_000, 80)

print(f"Volllaststunden Wind Onshore: {vlh_wind_onshore:.0f} h/a")
print(f"Volllaststunden Solar: {vlh_solar:.0f} h/a")


```

## 0.6 Klasse: Ein einfaches Kraftwerk

```python
# ============================================================
# OBJEKTORIENTIERUNG – Klasse als Bauplan
# ============================================================
# Eine Klasse ist ein Bauplan. Ein Objekt ist das fertige "Ding".
# Beispiel: Klasse = Kraftwerkstyp, Objekt = konkretes Kraftwerk

class Kraftwerk:
    """Modelliert ein einzelnes Kraftwerk mit Kosten und Erzeugung."""

    # ----------------------------------------------------------
    # __init__ = Konstruktor
    # Wird automatisch aufgerufen wenn man Kraftwerk(...) schreibt.
    # self = das Objekt selbst (wie "ich" aus Sicht des Objekts)
    # ----------------------------------------------------------
    def __init__(self, name: str, leistung_mw: float,
                 wirkungsgrad: float = 1.0,
                 brennstoffkosten: float = 0.0):
        self.name = name                  # Attribut: Bezeichnung
        self.leistung_mw = leistung_mw    # Attribut: Nennleistung in MW
        self.wirkungsgrad = wirkungsgrad  # Attribut: η = P_el / P_th
        self.brennstoffkosten = brennstoffkosten  # EUR/MWh_th

    # ----------------------------------------------------------
    # Methode = Funktion die zum Objekt gehört
    # Grenzkosten: wie teuer ist 1 MWh Strom aus diesem Kraftwerk?
    # Formel: Brennstoffkosten [EUR/MWh_th] / Wirkungsgrad [-]
    # ----------------------------------------------------------
    def grenzkosten(self) -> float:
        """Gibt variable Erzeugungskosten in EUR/MWh_el zurück."""
        return self.brennstoffkosten / self.wirkungsgrad

    # ----------------------------------------------------------
    # __repr__ = wie soll das Objekt als Text aussehen?
    # Wird bei print(objekt) aufgerufen.
    # ----------------------------------------------------------
    def __repr__(self):
        return f"Kraftwerk('{self.name}', P={self.leistung_mw} MW, η={self.wirkungsgrad:.0%})"


# ============================================================
# INSTANZIIERUNG – Objekte aus dem Bauplan erstellen
# ============================================================
# Jedes Objekt hat eigene Attributwerte, aber dieselben Methoden.
# Wind hat keine Brennstoffkosten → Grenzkosten = 0 EUR/MWh

windpark     = Kraftwerk("Windpark Nord", leistung_mw=150)
gaskraftwerk = Kraftwerk("GuD Süd",       leistung_mw=400,
                          wirkungsgrad=0.58, brennstoffkosten=35)
kohle        = Kraftwerk("Kohlekraftwerk", leistung_mw=800,
                          wirkungsgrad=0.40, brennstoffkosten=12)

print(windpark)       # ruft __repr__ auf
print(gaskraftwerk)
print(kohle)

# ============================================================
# MERIT ORDER – Methoden auf Objekten aufrufen
# ============================================================
# .grenzkosten() ist eine Methode – sie gehört zum Objekt
# und hat Zugriff auf alle Attribute über self.

kraftwerke = [windpark, gaskraftwerk, kohle]
# Kraftwerk.grenzkosten ist die Methode der Klasse – ohne Klammern,
# also nicht aufgerufen, sondern als Funktion übergeben.
# sorted() ruft sie intern für jedes Objekt auf
kraftwerke_sortiert = sorted(kraftwerke, key=Kraftwerk.grenzkosten)


print("\nMerit Order:")
for kw, counter in zip(kraftwerke_sortiert, range(1, len(kraftwerke_sortiert) + 1)):
    print(f"  {counter:2d}. {kw.name:20s}  →  {kw.grenzkosten():.2f} EUR/MWh")
```
