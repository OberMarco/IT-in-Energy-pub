### Beispiel 1: Einfacher Prompt – Funktion erstellen

**Prompt:**
> Schreibe eine Python-Funktion `volllaststunden`, die die Jahreserzeugung in MWh
> und die installierte Leistung in MW als Parameter nimmt und die Volllaststunden
> zurückgibt. Füge einen Docstring und Typannotationen hinzu. Validiere, dass
> die Leistung > 0 ist.

**Erwartetes Ergebnis (so oder ähnlich):**

````python
def volllaststunden(jahreserzeugung_mwh: float, installierte_leistung_mw: float) -> float:
    """
    Berechnet die Volllaststunden eines Kraftwerks.

    Args:
        jahreserzeugung_mwh: Jährliche Erzeugung in MWh
        installierte_leistung_mw: Installierte Leistung in MW

    Returns:
        Volllaststunden in h/a

    Raises:
        ValueError: Wenn Leistung <= 0
    """
    if installierte_leistung_mw <= 0:
        raise ValueError("Installierte Leistung muss > 0 sein!")
    return jahreserzeugung_mwh / installierte_leistung_mw


# Test
print(f"Wind Onshore:  {volllaststunden(500_000, 250):.0f} h/a")
print(f"PV Freifläche: {volllaststunden(90_000, 100):.0f} h/a")
````

### Beispiel 2: Mittlerer Prompt – Klasse erstellen

**Prompt:**
> Erstelle eine Python-Klasse `Batterie` mit folgenden Eigenschaften:
> - Attribute: kapazitaet_mwh, max_leistung_mw, aktueller_soc (0 bis 1), wirkungsgrad_laden, wirkungsgrad_entladen
> - Methode `laden(leistung_mw, dauer_h)`: Lädt die Batterie, berücksichtigt Wirkungsgrad und Kapazitätsgrenzen
> - Methode `entladen(leistung_mw, dauer_h)`: Entlädt die Batterie, gibt abgegebene Energie in MWh zurück
> - Methode `status()`: Gibt den aktuellen SOC in MWh und Prozent aus
> - Property `energie_mwh`: Gibt den aktuellen Energieinhalt in MWh zurück
>
> Verwende Typannotationen. Stelle sicher, dass der SOC nie unter 0 oder über 1 geht.

**Erwartetes Ergebnis:**

````python
class Batterie:
    """Einfaches Batteriespeicher-Modell."""

    def __init__(self, kapazitaet_mwh: float, max_leistung_mw: float,
                 soc: float = 0.5, eta_laden: float = 0.95,
                 eta_entladen: float = 0.95):
        self.kapazitaet_mwh = kapazitaet_mwh
        self.max_leistung_mw = max_leistung_mw
        self.soc = soc
        self.eta_laden = eta_laden
        self.eta_entladen = eta_entladen

    @property
    def energie_mwh(self) -> float:
        """Aktueller Energieinhalt in MWh."""
        return self.soc * self.kapazitaet_mwh

    def laden(self, leistung_mw: float, dauer_h: float = 1.0) -> float:
        """Lädt die Batterie. Gibt tatsächlich gespeicherte Energie in MWh zurück."""
        leistung_mw = min(leistung_mw, self.max_leistung_mw)
        energie_brutto = leistung_mw * dauer_h
        energie_netto = energie_brutto * self.eta_laden

        # Kapazitätsgrenze beachten
        platz_frei = (1 - self.soc) * self.kapazitaet_mwh
        energie_netto = min(energie_netto, platz_frei)

        self.soc += energie_netto / self.kapazitaet_mwh
        return energie_netto

    def entladen(self, leistung_mw: float, dauer_h: float = 1.0) -> float:
        """Entlädt die Batterie. Gibt abgegebene Energie in MWh zurück."""
        leistung_mw = min(leistung_mw, self.max_leistung_mw)
        energie_gewuenscht = leistung_mw * dauer_h

        # Verfügbare Energie prüfen
        verfuegbar = self.soc * self.kapazitaet_mwh
        energie_aus_speicher = min(energie_gewuenscht, verfuegbar)

        self.soc -= energie_aus_speicher / self.kapazitaet_mwh
        energie_abgegeben = energie_aus_speicher * self.eta_entladen
        return energie_abgegeben

    def status(self):
        """Gibt aktuellen Status aus."""
        print(f"SOC: {self.energie_mwh:.2f} MWh ({self.soc:.1%})")


# Test
bat = Batterie(kapazitaet_mwh=100, max_leistung_mw=50, soc=0.2)
bat.status()

gespeichert = bat.laden(50, dauer_h=1)
print(f"Geladen: {gespeichert:.2f} MWh")
bat.status()

abgegeben = bat.entladen(30, dauer_h=1)
print(f"Entladen: {abgegeben:.2f} MWh")
bat.status()
````

### Beispiel 3: Komplexer Prompt – Datenverarbeitung

**Prompt:**
> Ich habe eine pandas DataFrame `df` mit stündlichen Strompreisen in der Spalte 'price' (EUR/MWh)
> und einem DatetimeIndex. Schreibe Python-Code, der:
> 1. Den Durchschnittspreis pro Tag berechnet
> 2. Die Stunde mit dem höchsten und niedrigsten Preis pro Tag findet
> 3. Die Anzahl negativer Preisstunden pro Monat zählt
> 4. Das Ergebnis als übersichtliche Tabelle ausgibt
> Verwende pandas. Keine Schleifen, nutze Gruppierungen.

````python
import pandas as pd
import numpy as np

# Beispieldaten erstellen (7 Tage stündlich)
np.random.seed(42)
idx = pd.date_range("2024-01-01", periods=168, freq="h")
preise = 50 + 30 * np.sin(np.arange(168) * 2 * np.pi / 24) + np.random.normal(0, 15, 168)
df_demo = pd.DataFrame({"price": preise}, index=idx)

# 1. Durchschnittspreis pro Tag
tagespreise = df_demo.resample("D")["price"].mean()
print("Durchschnittspreis pro Tag:")
print(tagespreise.round(2))

# 2. Stunde mit höchstem/niedrigstem Preis pro Tag
print("\nPreisextreme pro Tag:")
for tag, gruppe in df_demo.groupby(df_demo.index.date):
    max_idx = gruppe["price"].idxmax()
    min_idx = gruppe["price"].idxmin()
    print(f"  {tag}: Max {gruppe['price'].max():.1f} EUR um {max_idx.hour}:00, "
          f"Min {gruppe['price'].min():.1f} EUR um {min_idx.hour}:00")

# 3. Negative Preisstunden pro Monat
negative = df_demo[df_demo["price"] < 0].resample("ME").size()
print(f"\nNegative Preisstunden pro Monat: {dict(negative)}")
````

## 2.5 Grenzen und Risiken von KI-generiertem Code

| Risiko | Beispiel |
|--------|---------|
| **Halluzinationen** | KI erfindet Funktionen, die es nicht gibt |
| **Veralteter Code** | Nutzt deprecated APIs |
| **Subtile Fehler** | Code läuft, aber Ergebnis ist falsch (z.B. Einheitenfehler MW ↔ kW) |
| **Sicherheit** | Unsichere Passwort-Handhabung, SQL-Injection |
| **Lizenzprobleme** | Generierter Code könnte aus GPL-Quellen stammen |


> ⚠️ **Goldene Regel:** Code, den du nicht verstehst, solltest du nicht verwenden!

### Aufgabe 3: Debugging mit KI

Der folgende Code hat Fehler. Kopiere ihn in ein KI-Tool mit dem Prompt:
> Finde und korrigiere die Fehler in diesem Python-Code. Erkläre jeden Fehler.



````python
# Fehlerhafter Code – zur Analyse an KI geben:
def berechne_lcoe(investition, betriebskosten_jahr, erzeugung_jahr_mwh, laufzeit, zinssatz):
    """Berechnet Levelized Cost of Energy"""
    # Fehler 1: Barwertberechnung falsch
    barwert_kosten = investition
    for jahr in range(laufzeit):
        barwert_kosten += betriebskosten_jahr / (1 + zinssatz) * jahr

    # Fehler 2: Barwert der Erzeugung falsch
    barwert_erzeugung = erzeugung_jahr_mwh * laufzeit

    # Fehler 3: Division
    lcoe = barwert_erzeugung / barwert_kosten

    return lcoe

# Test (sollte ca. 50-80 EUR/MWh für Wind ergeben)
ergebnis = berechne_lcoe(
    investition=1_500_000,      # EUR
    betriebskosten_jahr=50_000, # EUR/a
    erzeugung_jahr_mwh=5_000,   # MWh/a
    laufzeit=20,                # Jahre
    zinssatz=0.05               # 5%
)
print(f"LCOE: {ergebnis:.2f} EUR/MWh")
````
